Realice una función que dado un arreglo que calcule la media, el valor mínimo y el valor máximo del mismo arreglo.

Media: es el número "promedio"; se encuentra al sumar todos los datos y dividir entre el número de datos.
image.png

Valor mínimo: es el valor más pequeño entre todas las muestras registradas.

Valor máximo: es el valor más grande entre todas las muestras registradas.




EJEMPLO:

```python
#definimos nuestras vaariables
lista = (3,5,2.5,3.6,17,6.6,4.5,34,56,12,11.11,26,234.5,12.65,1)
```

```python
#hacemos las operaciones de la media aritmetica donde len va a devolver la longitud de la lista
media = sum(lista) / len(lista)
```

```python
#finalmente nos muestra el resultado de la media
print (media)
```

```python
#Para mostrar el valor minimo de la lista utilizamos la funcion min
min(lista)
```

```python
#Para mostrar el valor maximo de la lista utilizamos la funcion max
max(lista)
```
